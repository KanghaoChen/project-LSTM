```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
import pandas as pd
path = os.listdir('outputsdata')
balanced = pd.read_excel('outputsdata/'+path[0])
precision = pd.read_excel('outputsdata/'+path[1])
recall = pd.read_excel('outputsdata/'+path[2])
Roc = pd.read_excel('outputsdata/'+path[3])
trainloss = pd.read_excel('outputsdata/'+path[4])
validloss = pd.read_excel('outputsdata/'+path[5])
```

```python
def reset_index(df):
    return df.set_index('Step',drop=True)
```

```python
balanced = reset_index(balanced)
precision = reset_index(precision)
recall = reset_index(recall)
Roc = reset_index(Roc)
trainloss = reset_index(trainloss)
validloss = reset_index(validloss)
```

```python
from scipy.interpolate import make_interp_spline, BSpline
```

```python
x = np.array(balanced.index)
y = np.array(balanced[balanced.columns])
xnew = np.linspace(x.min(), x.max(), 200) 

#define spline
spl = make_interp_spline(x, y, k=2)
y_smooth = spl(xnew)

#create smooth line chart 
plt.plot(xnew, y_smooth)
plt.show()
```

```python
balanced.plot(kind='line', title='Balanced Accuracy')
```
